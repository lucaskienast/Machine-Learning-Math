# Introduction to Partial Derivatives

```python
import numpy as np
import matplotlib.pyplot as plt
```

Consider the function $f(x, y)$ for $z = x^2 - y^2$:

```python
def f(my_x, my_y):
  return my_x**2 - my_y**2
```

## z wrt. x

```python
xs = np.linspace(-3, 3, 1000)
```

Keep $y$ constant (e.g., at $y = 0$):

```python
zs_wrt_x = f(xs, 0)
```

```python
fig, ax = plt.subplots()
plt.axvline(x=0, color="lightgray")
plt.axhline(y=0, color="lightgray")
plt.xlabel("x")
plt.ylabel("z", rotation=0)
_ = ax.plot(xs, zs_wrt_x)
```

To determine the slope of $z$ w.r.t. $x$ at a given point along the curve, we use the partial derivative:$ \frac{\partial z}{\partial x} = 2x$

```python
def delz_delx(my_x, my_y): # y not relevant here, but can be in other examples
  return 2*my_x
```

```python
x_samples = [-2, -1, 0, 1, 2]
```

```python
colors = ["red", "orange", "green", "blue", "magenta"]
```

```python
def plot_point_and_tangent_wrt_x(my_xs, my_x, my_y, my_f, fprime, col):
  
  my_z = my_f(my_x, my_y)
  plt.scatter(my_x, my_z, c=col, zorder=3)

  tangent_m = fprime(my_x, my_y)
  tangent_b = my_z - tangent_m*my_x
  tangent_line = tangent_m*my_xs + tangent_b

  plt.plot(my_xs, tangent_line, c=col, linestyle="--", linewidth=0.7, zorder=3)
```

```python
fig, ax = plt.subplots()
plt.axvline(x=0, color="lightgray")
plt.axhline(y=0, color="lightgray")

for i, x in enumerate(x_samples):
  plot_point_and_tangent_wrt_x(xs, x, 0, f, delz_delx, colors[i])

plt.ylim(-1,9)
plt.xlabel("x")
plt.ylabel("z", rotation=0)
_ = ax.plot(xs, zs_wrt_x)
```

The slope of $z$ along the $x$ axis is twice the corresponding $x$ value.

## z wrt. y

```python
ys = np.linspace(-3, 3, 1000)
```

Keep $x$ constant (e.g., at $x = 0$):

```python
zs_wrt_y = f(0, ys)
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(12,5))
plt.axvline(x=0, color='lightgray')
plt.axhline(y=0, color='lightgray')

plt.xlabel('y')
plt.ylabel('z', rotation=0)

_ = ax[0].plot(ys, zs_wrt_y) # f(0, ys)
_ = ax[1].plot(ys, f(2, ys)) # z = x^2 + y^2 i.e. at y=0: z = 2^2 + 0 = 4
```

The partial derivative of $z$ w.r.t $y$ is independent of $x$ (just as we observed $x$ is independent of $y$ above), so while $z$ varies as a function of both $x$ and $y$, the slope of $z$ w.r.t $y$ is the same no matter what $x$ is: $ \frac{\partial z}{\partial y} = -2y $

So for example, holding $x$ constant at 2 instead of 0 increases $z$, but has no impact whatsoever on the slope of $z$ w.r.t. $y$:

```python
def delz_dely(my_x, my_y):
    return -2*my_y
```

```python
y_samples = [-2, -1, 0, 1, 2]
```

```python
def plot_point_and_tangent_wrt_y(my_ys, my_x, my_y, my_f, fprime, col):
  
  my_z = my_f(my_x, my_y)
  plt.scatter(my_y, my_z, c=col, zorder=3)

  tangent_m = fprime(my_x, my_y)
  tangent_b = my_z - tangent_m*my_y
  tangent_line = tangent_m*my_ys + tangent_b

  plt.plot(my_ys, tangent_line, c=col, linestyle="--", linewidth=0.7, zorder=3)
```

```python
fig, ax = plt.subplots()
plt.axvline(x=0, color='lightgray')
plt.axhline(y=0, color='lightgray')

for i, y in enumerate(y_samples):
    plot_point_and_tangent_wrt_y(ys, 2, y, f, delz_dely, colors[i])

plt.ylim(-5, 5)
plt.xlabel('y')
plt.ylabel('z', rotation=0)

zs_wrt_y = f(2, ys)
_ = ax.plot(xs, zs_wrt_y)
```

The slope of $z$ along the $y$ axis is twice the $y$ value and inverted, resulting in the parabolic curve opening downward.
